Treine um modelo do tp DRF, utilizando as config default, c um Train-Test split de 75%-25%. Avalie a qualidade do modelo tendo em atenção as métricas abordadas em contexto de aula.

```python
# pip install pandas scikit-learn h2o
import pandas as pd
import h2o
from sklearn.model_selection import train_test_split
from h2o.estimators import H2ORandomForestEstimator
```

```python
# Inicializa o sv H2O.
# Isso config um ambiente H2O local pra executar modelos de ML.
h2o.init()

# Carregar o dataset
data = h2o.import_file('mushrooms.csv', header=1)

data
```

```python
# Divide aleatoriamente os dados em 2 conj disjuntos: treino e teste.
train, test = data.split_frame(ratios=[0.75], seed=123)
```

```python
# Treina um modelo de DRF

# Config e treinar o modelo DRF c config default
model = H2ORandomForestEstimator()

# Treina o modelo
model.train(x=list(train.columns[1:]), y="class", training_frame=train)
```

```python
# As métricas de desempenho comuns serão calculadas no conj de teste.
performance_default = model.model_performance(test_data=test)
print(performance_default)
```

Treine 2 modelos adicionais do mm tp, mas alterando algumas das suas config.

```python
# Modelo 1: + trees
drf_more_trees = H2ORandomForestEstimator(ntrees=100)
drf_more_trees.train(x=list(train.columns[1:]), y="class", training_frame=train)

# Modelo 2: Profundidade Máx Aumentada
drf_deeper = H2ORandomForestEstimator(ntrees=50, max_depth=30)
drf_deeper.train(x=list(train.columns[1:]), y="class", training_frame=train)

performance_more_trees = drf_more_trees.model_performance(test_data=test)
performance_deeper = drf_deeper.model_performance(test_data=test)

print("Desempenho do Modelo c + trees:")
print(performance_more_trees)

print("Desempenho do Modelo c > Profundidade:")
print(performance_deeper)
```

Compare, de forma crítica, os 3 modelos treinados e escolha, de forma fundamentada, o mlhr modelo pra eventualmente ser colocado em produção.

```python
# Exibir as métricas chave para tomar a decisão
print("Métricas do Modelo default:")
print(performance_default.auc(), performance_default.accuracy())

print("Métricas do Modelo c + trees:")
print(performance_more_trees.auc(), performance_more_trees.accuracy())

print("Métricas do Modelo c > Profundidade:")
print(performance_deeper.auc(), performance_deeper.accuracy())
```
